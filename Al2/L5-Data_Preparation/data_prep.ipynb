```python
gps_data = pd.read_csv('googleplaystore.csv')
gps_data.drop(index=[10472], inplace=True)
gps_data.dropna(subset=['Rating'], inplace=True)
gps_data.info()
```

```python
# Check for unique values in column Category
gps_data['Category'].unique()
```

Encode the Category column using LabelEncoder

1.   Import and declare encoder as category_le
2.   Fit category_le with gps_data['Category']
3.   List out the classes that category_le can encode
4.   Encode the 'Category' column of gps_data and reassign it back to gps_data['Category']

```python
# Import and declare encoder
from sklearn.preprocessing import LabelEncoder

category_le = LabelEncoder()  

# Fit encoder with gps_data['Category']
category_le.fit(gps_data['Category'])

# List out the classes in which category_le can encode
category_le.classes_
```

```python
# Encode the 'Category' column of gps_data and reassign it back to gps_data['Category']
gps_data['Category'] = category_le.fit_transform(gps_data['Category'])
```
