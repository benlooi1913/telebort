```python
import cv2
```

```python
# Load trained data 
trained_face_data = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
trained_face_data = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye_tree_eyeglasses.xml")
```

```python
img = cv2.imread("The Rock.jpeg")
if img is None:
    print("Error: Image 'The Rock.jpeg' not found or cannot be loaded.")
else:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face_coordinates = trained_face_data.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5
    )
    # Draw rectangles around faces
    for (x, y, w, h) in face_coordinates:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

    for i, (x, y, w, h) in enumerate(face_coordinates, start=1):
         cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
         cv2.putText(img, f"Face {i}", (x, y-10),
                     cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    cv2.imshow("Face Detection", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
